```python
L2f = []
L2b = []
L2c = []
n = 10
Ns = []

while n < 6000:
    
    x = np.linspace(0,10,n)
    f = np.sin(x)

    dx = x[1] - x[0]

    dfdx_f = (f[2:] - f[1:-1])/dx

    dfdx_b = (f[1:-1] - f[:-2])/dx

    dfdx_c = (f[2:] - f[:-2]) / (2*dx)
    dfdx = np.cos(x[1:-1])

    L2f.append(np.sqrt(np.sum((dfdx_f - dfdx)**2)))
    L2b.append(np.sqrt(np.sum((dfdx_b - dfdx)**2)))
    L2c.append(np.sqrt(np.sum((dfdx_c - dfdx)**2)))
    
    Ns.append(n)
    n *= 1.5    
    n = int(n)
```

```python
fig, ax = pyp.subplots(figsize = (6,6))

ax.plot(Ns, L2f, label = 'forward', marker = 'X', color = 'dodgerblue')
ax.plot(Ns, L2b, label = 'backward', marker = 'H', color='orangered')
ax.plot(Ns, L2c, label = 'central', marker = 'o', color = '#66AF00')

# ax.plot(Ns, 1/np.array(Ns), color = 'k', label = "First Order Slope", linestyle = ":")
# ax.plot(Ns, .5/np.array(Ns), color = 'k', label = "First Order Slope", linestyle = ":")


ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')

ax.set_ylabel("L2 norm")
ax.set_xlabel("Number of Points")

ax.grid(True)
```
